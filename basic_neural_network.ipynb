```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def deriv_sigmoid(x):
    fx = sigmoid(x)
    return fx * (1 - fx)

def mse(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

class basicNeuralNetwork:
    """ 
    - 2 inputs
    - 1 hidden layer with 2 neurons
    - 1 output layer with 1 neuron
    """
    def __init__(self):
        #w
        self.w1 = np.random.normal()
        self.w2 = np.random.normal()
        self.w3 = np.random.normal()
        self.w4 = np.random.normal()
        self.w5 = np.random.normal()
        self.w6 = np.random.normal()

        #b
        self.b1 = np.random.normal()
        self.b2 = np.random.normal()
        self.b3 = np.random.normal()

    def feedforward(self, x):
        h1 = sigmoid(self.w1 * x[0] + self.w2 * x[1] + self.b1)
        h2 = sigmoid(self.w3 * x[0] + self.w4 * x[1] + self.b2)
        o1 = sigmoid(self.w5 * h1 + self.w6 * h2 + self.b3)
        return o1
    
    def train(self, data, all_y_trues):
        lr = 0.1
        epoches = 1000


        for epoch in range(epoches):
            for x, y_true in zip(data, all_y_trues):
                sum_h1 = self.w1 * x[0] + self.w2 * x[1] + self.b1
                h1 = sigmoid(sum_h1)

                sum_h2 = self.w3 * x[0] + self.w4 * x[1] + self.b2
                h2 = sigmoid(sum_h2)

                sum_o1 = self.w5 * h1 + self.w6 * h2 + self.b3
                o1 = sigmoid(sum_o1)
                y_pred = o1

                d_L_d_ypred = -2 * (y_true - y_pred)

                #o1
                d_ypred_d_w5 = h1 * deriv_sigmoid(sum_o1)
                d_ypred_d_w6 = h2 * deriv_sigmoid(sum_o1)
                d_ypred_d_b3 = deriv_sigmoid(sum_o1)

                #h1
                d_ypred_d_w1 = self.w5 * deriv_sigmoid(sum_o1) * x[0] * deriv_sigmoid(sum_h1)
                d_ypred_d_w2 = self.w5 * deriv_sigmoid(sum_o1) * x[1] * deriv_sigmoid(sum_h1)
                d_ypred_d_b1 = self.w5 * deriv_sigmoid(sum_o1) * deriv_sigmoid(sum_h1)

                #h2
                d_ypred_d_w3 = self.w6 * deriv_sigmoid(sum_o1) * x[0] * deriv_sigmoid(sum_h2)
                d_ypred_d_w4 = self.w6 * deriv_sigmoid(sum_o1) * x[1] * deriv_sigmoid(sum_h2)
                d_ypred_d_b2 = self.w6 * deriv_sigmoid(sum_o1) * deriv_sigmoid(sum_h1)

                #update
                self.w1 -= lr * d_L_d_ypred * d_ypred_d_w1
                self.w2 -= lr * d_L_d_ypred * d_ypred_d_w2
                self.w3 -= lr * d_L_d_ypred * d_ypred_d_w3
                self.w4 -= lr * d_L_d_ypred * d_ypred_d_w4
                self.w5 -= lr * d_L_d_ypred * d_ypred_d_w5
                self.w6 -= lr * d_L_d_ypred * d_ypred_d_w6
                self.b1 -= lr * d_L_d_ypred * d_ypred_d_b1
                self.b2 -= lr * d_L_d_ypred * d_ypred_d_b2
                self.b3 -= lr * d_L_d_ypred * d_ypred_d_b3

                if epoch % 20 == 0:
                    y_preds = np.apply_along_axis(self.feedforward, 1, data)
                    loss = mse(y_preds, all_y_trues)
                    print("Epoch %d loss: %.3f" % (epoch,loss))


data = np.array([
  [-2, -1],  # Alice
  [25, 6],   # Bob
  [17, 4],   # Charlie
  [-15, -6], # Diana
])
all_y_trues = np.array([
  1, # Alice
  0, # Bob
  0, # Charlie
  1, # Diana
])

network = basicNeuralNetwork()
network.train(data, all_y_trues)
```

```python
emily = np.array([-7, -3]) # 128 pounds, 63 inches
frank = np.array([20, 2])  # 155 pounds, 68 inches
print("Emily: %.3f" % network.feedforward(emily)) # 0.951 - F
print("Frank: %.3f" % network.feedforward(frank)) # 0.039 - M
```
